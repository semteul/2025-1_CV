## Matplotlib 기본 활용법

Matplotlib: 다양한 그래프 및 다이어그램을 그리는 도구를 지원하는 라이브러리

### 기본적인 그래프

- plt.figure(num=None, figsize=None, ...) - 새로운 figure 생성
- plt.plot(* args, ..., ** kwargs) / args: [x], y, [fmt] - x에 대한 y의 선형 그래프. 이때 x와 y의 크기가 같아야함
- plt.show() - 모든 figure를 display

```python
x = np.arange(50)
y = np.arange(50)
```

```python
x, y
```

#### 기본 ploting

```python
plt.figure(figsize=(4,4))
plt.plot(x, y)
plt.show()
```

#### 여러 figure ploting

```python
y2 = np.arange(50, 100)
```

```python

plt.figure(figsize=[5, 5])
plt.plot(x, y)
plt.plot(x, y2)
plt.show()

plt.figure(figsize=[5, 5])
plt.plot(x, y2)
plt.show()
```

#### Subplot의 활용

```python
fig, axes = plt.subplots(2, 2, figsize=[8, 8])
axes[0, 0].plot(x, y)
axes[0, 1].plot(x, y*y)
axes[1, 0].plot(x, y**3)
axes[1, 1].plot(x, -y)
plt.show()
```

#### Plot 옵션

```python
fig, axes = plt.subplots(2,2, figsize=[8, 8])
axes[0, 0].plot(x, y)        # plot x and y using default line style and color
axes[0, 1].plot(x, y, 'bo')  # plot x and y using blue circle markers
axes[1, 0].plot(y)           # plot y using x as index array 0..N-1
axes[1, 1].plot(y, 'r+')     # ditto, but with red plusses
plt.show()
```

#### Draw Heatmap

```python
numeric_data = data.select_dtypes(include=np.number)
corr = numeric_data.corr()

plt.figure(figsize = [7, 3])
sns.heatmap(corr, cmap = 'Purples', annot = True)
```

### Quantile plot

```python
from scipy.stats import probplot
```

```python
fig, axes = plt.subplots(2,2, figsize=[10, 10])

probplot(data['Speed'], plot=axes[0, 0])
axes[0, 0].set_title('Speed')
probplot(data['Attack'], plot=axes[0, 1])
axes[0, 1].set_title('Attack')
probplot(data['Defense'], plot=axes[1, 0])
axes[1, 0].set_title('Defense')
plt.show()
```

```python
train = pd.read_csv(basicpath + 'biketrain.csv', parse_dates=['datetime'])
```

```python
train.head()
```

```python
train['datetime-year'] = train['datetime'].dt.year
train['datetime-month'] = train['datetime'].dt.month
train['datetime-day'] = train['datetime'].dt.day
train['datetime-hour'] = train['datetime'].dt.hour
train['datetime-minute'] = train['datetime'].dt.minute
train['datetime-second'] = train['datetime'].dt.second
```

```python
train['datetime'].dt.month
```

```python
train.head()
```

matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)

```python
# 24시간 동안 자전거 count의 변화
x = train[:24]['datetime-hour']
y = train[:24]['count']

plt.plot(x, y)
```

seaborn.pointplot(*, x=None, y=None, hue=None, data=None, order=None, hue_order=None, estimator=<function mean at 0x7ff320f315e0>, ci=95, n_boot=1000, units=None, seed=None, markers='o', linestyles='-', dodge=False, join=True, scale=1, orient=None, color=None, palette=None, errwidth=None, capsize=None, ax=None, **kwargs)

```python

#24시간 동안 각 시각별로 전체 자전거 대여 횟수 평균의 변화
sns.pointplot(data = train, x = 'datetime-hour', y = 'count')

```

Quiz. 주말과 주중의 자전거 대여 횟수 plot을 하나의 canvas에 각각 그리시오

```python
fig, axes = plt.subplots(2, 1, figsize=[10, 10])

axes[0].set_title('workingday')
sns.pointplot(data=train.loc[train.workingday == 1], x='datetime-hour', y='count', ax=axes[0])

axes[1].set_title('not workingday')
sns.pointplot(data=train.loc[train.workingday == 0], x='datetime-hour', y='count', ax=axes[1])

#sns.barplot(data = result, x='workingday', y='count')
```
